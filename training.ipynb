```python
# Install Ultralytics first
# pip install ultralytics

from ultralytics import YOLO

# Load pretrained YOLOv8 model
model = YOLO("yolov8m.pt")   # n=nano, s=small, m=medium, l=large, x=xlarge

# Train the model
model.train(
    data="/content/kidney-2/data.yaml",     # path to dataset config file
    epochs=100,           # number of training epochs
    imgsz=640,            # image size
    batch=16,             # batch size
    device=0,             # GPU id (0) or "cpu"
    workers=4,            # dataloader workers
    project="runs/train", # save folder
    name="kidney"           # experiment name
)

model.eval()
```
